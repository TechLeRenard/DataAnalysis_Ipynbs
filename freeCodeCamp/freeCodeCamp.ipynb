```python
import pandas as pd


def calculate_demographic_data(print_data=True):
    # Read data from file
    df = pd.read_csv("adult.data.csv")

    # How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.
    race_count = pd.Series(df["race"].value_counts())

    # What is the average age of men?
    average_age_men = round(df.loc[df['sex'] == 'Male', 'age'].mean(),1)

    # What is the percentage of people who have a Bachelor's degree?
    percentage_bachelors=df.loc[df['education'] == 'Bachelors', 'education'].value_counts()/df["education"].value_counts().sum()
    percentage_bachelors = round(float(percentage_bachelors*100),1)

    # What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
    # What percentage of people without advanced education make more than 50K?

    # with and without `Bachelors`, `Masters`, or `Doctorate`
    ##### I get the rows, under the 'Education', who has 'Bachelors' or 'Masters' or 'Doctorate'
    higher_education = df.loc[(df['education'] == 'Bachelors') | (df['education'] == 'Masters') | (df['education'] == 'Doctorate'),'education'].value_counts().sum()
    ##### I get the rows, under the 'Education', who is not 'Bachelors' AND not 'Masters' AND not 'Doctorate'
    lower_education = df.loc[(df['education'] != 'Bachelors') & (df['education'] != 'Masters') & (df['education'] != 'Doctorate'),'education'].value_counts().sum()

    # percentage with salary >50K
    #### in my understanding, you get the result filtered(== >50K) value_counts.sum of higher_education over 
    ####the population of higher ed
    # same goes to lower_education_rich
    
    higher_Bach = df.loc[(df['education'] == 'Bachelors') & (df['salary'] == '>50K'),'education'].value_counts().sum()
    higher_Mast = df.loc[(df['education'] == 'Masters') & (df['salary'] == '>50K'),'education'].value_counts().sum()
    higher_Doc = df.loc[(df['education'] == 'Doctorate') & (df['salary'] == '>50K'),'education'].value_counts().sum()
    total_higher_rich = higher_Bach + higher_Mast + higher_Doc

    higher_education_rich = total_higher_rich/higher_education
    higher_education_rich = round(higher_education_rich*100,1)
    lower_education_rich = df.loc[(df['education'] != 'Bachelors') & (df['education'] != 'Masters') & (df['education'] != 'Doctorate') & (df['salary'] == '>50K'),'education'].value_counts().sum()/lower_education
    lower_education_rich = round(lower_education_rich*100,1)
    # What is the minimum number of hours a person works per week (hours-per-week feature)?
    min_work_hours = df["hours-per-week"].min()

    # What percentage of the people who work the minimum number of hours per week have a salary of >50K?
    num_min_workers = df.loc[df['hours-per-week'] == min_work_hours, 'hours-per-week'].value_counts().sum()

    rich_min = df.loc[(df['hours-per-week'] == min_work_hours) & (df['salary'] == ">50K"),'hours-per-week'].value_counts().sum()
    rich_percentage = rich_min/num_min_workers*100
    
    # What country has the highest percentage of people that earn >50K?
    ##### So, I filtered the (col native_country who has >50k) then divided over (total population of that country)
    ##### fortunately, the division process was automatically aligned which matches it by the index, which is the country name
    ##### in analogy number of rich person in that country divided by total person in that country * 100
    get_whole_list = ((df.loc[df['salary'] == ">50K", 'native-country'].value_counts())/(df['native-country'].value_counts()))*100
    get_whole_list = get_whole_list.sort_values(ascending=False)
    highest_earning_country = get_whole_list.index[0]
    highest_earning_country_percentage = round(get_whole_list[0],1)

    # Identify the most popular occupation for those who earn >50K in India.
    #### I created another df which constains all cols but country is only India
    
    
    ind_row = df.loc[df['native-country'] == "India", 'native-country']
    ind_df = df.iloc[ind_row.index]
    #### Then filter again since we focus only in occupation having above 50K
    occ_in = ind_df.loc[ind_df['salary'] == ">50K", 'occupation'].value_counts()
    occ_in = str(occ_in.index[0])
    
    top_IN_occupation = occ_in

    # DO NOT MODIFY BELOW THIS LINE

    if print_data:
        print("Number of each race:\n", race_count) 
        print("Average age of men:", average_age_men)
        print(f"Percentage with Bachelors degrees: {percentage_bachelors}%")
        print(f"Percentage with higher education that earn >50K: {higher_education_rich}%")
        print(f"Percentage without higher education that earn >50K: {lower_education_rich}%")
        print(f"Min work time: {min_work_hours} hours/week")
        print(f"Percentage of rich among those who work fewest hours: {rich_percentage}%")
        print("Country with highest percentage of rich:", highest_earning_country)
        print(f"Highest percentage of rich people in country: {highest_earning_country_percentage}%")
        print("Top occupations in India:", top_IN_occupation)

    return {
        'race_count': race_count,
        'average_age_men': average_age_men,
        'percentage_bachelors': percentage_bachelors,
        'higher_education_rich': higher_education_rich,
        'lower_education_rich': lower_education_rich,
        'min_work_hours': min_work_hours,
        'rich_percentage': rich_percentage,
        'highest_earning_country': highest_earning_country,
        'highest_earning_country_percentage':
        highest_earning_country_percentage,
        'top_IN_occupation': top_IN_occupation
    }

calculate_demographic_data()
```
